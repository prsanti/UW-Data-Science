```python
# Motor Vehicle Collision Data
df_vehicle_collision = pd.read_csv('Motor Vehicle Collisions with KSI Data - 4326.csv')

# Drop Columns
df_vehicle_collision.drop(columns=['_id', 'OFFSET', 'STREET2', 'DIVISION', 'ACCLOC', 'RDSFCOND', 'RDSFCOND', 'IMPACTYPE', 'INVTYPE', 'FATAL_NO', 'INITDIR', 'VEHTYPE', 'MANOEUVER', 'DRIVACT', 'PEDTYPE', 'PEDCOND', 'CYCACT', 'CYCCOND', 'PASSENGER', 'POLICE_DIVISION', 'HOOD_ID', 'ObjectId', 'geometry'],
                          axis=1,
                          inplace=True)

# convert date to datetime formate
df_vehicle_collision['DATE'] = pd.to_datetime(df_vehicle_collision['DATE'], format='%Y-%m-%d')

# filter date starting at 2015
df_vehicle_collision = df_vehicle_collision.loc[(df_vehicle_collision['DATE'] >= '2015-01-01')]


df_vehicle_collision
```

```python
# Time series of injuries every quarter
df_date_injury = df_vehicle_collision[['DATE', 'INJURY']]

# set date to index
df_date_injury = df_date_injury.set_index('DATE')

# Convert Fatal to 1
def replace_fatal_values(injury):
    # 0 - Non-fatal
    # 1 - Fatal
    if (injury == 'None' or injury == 'Minimal' or injury == 'Minor' or injury == 'Major'):
        return 0
    elif (injury == 'Fatal'):
        return 1


df_fatal = df_date_injury.copy()
df_fatal['INJURY'] = df_fatal['INJURY'].apply(replace_fatal_values)
df_fatal.rename(columns={"INJURY": "FATAL INJURIES"}, inplace=True)
# Resample by Quarterly and Sum
df_fatal = df_fatal.resample('Q').sum()
df_fatal.plot(title="Fatal Injuries per Quarter", xlabel="Yearly Quarter", ylabel="Fatal Injury Count", figsize=(20,10))
    
# Convert Non-Fatal to 1
def replace_nonfatal_values(injury):
    # 1 - Non-fatal
    # 0 - Fatal
    if (injury == 'None' or injury == 'Minimal' or injury == 'Minor' or injury == 'Major'):
        return 1
    elif (injury == 'Fatal'):
        return 0
    
df_nonfatal = df_date_injury.copy()
df_nonfatal['INJURY'] = df_nonfatal['INJURY'].apply(replace_nonfatal_values)
df_nonfatal.rename(columns={"INJURY": "NONFATAL INJURIES"}, inplace=True)
# Resample by Quarterly and Sum
df_nonfatal = df_nonfatal.resample('Q').sum()
df_nonfatal.plot(title="Non-Fatal Injuries per Quarter", xlabel="Yearly Quarter", ylabel="Non-Fatal Injury Count", figsize=(20,10))

ax = df_fatal.plot()
df_nonfatal.plot(ax=ax, title="Injuries per Quarter", xlabel="Quarter", ylabel="Injury Count", figsize=(20,10))
```

## Analysis - Seasonal Fatal vs Non-Fatal Injuries

Non-fatal injuries follow a trend where the number of non-fatal injuries start low in the winter, increases over time and peaks during the summer for every year. However, in 2020 after winter non-fatal injuries decreases in the spring probably due to covid since people started to work from home and drove less.

Fatal injuries appear to be more irregular. However, it follows the trend of non-fatal injuries in 2019.

```python
# Comparing time of day to injury type
df_light_injury = df_vehicle_collision[['DATE', 'LIGHT', 'INJURY']]
df_light_injury.set_index('INJURY', inplace=True)
#df_light_injury

#light_injury_count = df_light_injury.groupby('LIGHT')['INJURY'].value_counts()

light_injury_count = df_light_injury.groupby('INJURY')['LIGHT'].value_counts()
print(light_injury_count)
light_injury_plot = light_injury_count.plot(kind = "bar", title="Injuries", xlabel="Injury Severity at Light Condition", ylabel="Injury Count", figsize=(25,15))

# % of Fatal Injuries during Daylight to All Fatal Injuries
fatal_daylight = light_injury_count['Fatal']['Daylight']
total_fatal = light_injury_count['Fatal'].sum()
print("% of Fatal Injuries During Daylight: ", fatal_daylight / total_fatal)
```

## Analysis - Time of Day

The most common time of day when accidents oocur is during daylight, for all types of accidents. Since there are more drivers during the day, there are more accidents. The next common time of day when accidents occur is when it's Dark (artificial). This could be because drivers are more tired at night, and not fully concentrating on driving. The least common time for injuries (excluding other) is at dusk (artificial).

The most common type of accidents are when there are no injuries, then major injuries. Minor, minimal, and fatal injuries have about the same amount of occurence.

Analyzing only fatal injuries, 50.79% of all fatal injuries occur when there is daylight. About half of the fatal injuries occur during daytime maybe because there are more drivers, cyclists, and pedestrians on the road.

```python
# Comparing speeding to injury type
df_speeding_injury = df_vehicle_collision[['DATE', 'SPEEDING', 'INJURY']]
df_speeding_injury.set_index('INJURY', inplace=True)

speed_injury_count = df_speeding_injury.groupby('INJURY')['SPEEDING'].value_counts()
print(speed_injury_count)
speed_injury_plot = speed_injury_count.plot(kind = "bar", title="Injuries", xlabel="Injury Severity", ylabel="Injury Count", figsize=(25,15))

# % of Fatal Injuries of Speeding
total_fatal = speed_injury_count['Fatal']
total_speeding = speed_injury_count.sum()
print("% of Fatal Injuries due to Speeding: ", total_fatal / total_speeding)
```

## Analysis - Speeding

Speeding causes the most major injuries with 296 counts, then no injuries at 281. Speeding does not cause a significant amount of minimal, minor, and fatal injuries compared to major and no injuries.

The percentage of fatal injuries from speeding is about 9.18% and non-fatal injuries from speeding occur at 90.82%.
