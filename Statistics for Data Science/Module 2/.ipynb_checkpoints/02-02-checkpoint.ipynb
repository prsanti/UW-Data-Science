# Probability Distributions

A **probability distribution** is a list of possible outcomes with corresponding probabilities that satisfy three rules:

- **Rule 1**: The outcomes must be disjoint (mutually exclusive).
  
  
- **Rule 2**: Each probability must be between 0 and 1.
  
  
- **Rule 3**: The probabilities must total 1 (collectively exhaustive).

# Multiple Random Processes

## Independent events

Just as variables and observations can be independent, random processes can be independent as well. Two processes are independent if knowing the outcome of one provides no useful information about the outcome of the other.

For instance, flipping a coin and rolling a die are two independent processes &mdash; knowing the coin was heads does not help determine the outcome of a die roll. On the other hand, stock prices usually move up or down together, so they are not independent.

## Multiplication rule for independent processes

If $A$ and $B$ represent events from two different and independent processes, then the probability that both $A$ and $B$ occur can be calculated as the product of their separate probabilities:

$$P(A \ and \ B) = P(A\cap B)=P(A) \cdot P(B)$$

Similarly, if there are $k$ events $A_1$, ..., $A_k$ from $k$ independent processes, then the probability they all occur is:

$$
P(A_1 \ and \ A_2 \ ... \ and \ A_k) = P(A_1\cap A_2 ... \ \cap A_k)=P(A_1) \cdot P(A_2) \ ... \  \cdot P(A_k)
$$

Let's simulate two independent events: rolling two dice, and then calculate the probability that both dice yield a 6.

```python
import random
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
# This will ignore the warnings
import warnings
warnings.filterwarnings("ignore")

dice_rolls = np.arange(1,7)
count_hits = 0
x_data = []
y_data = []

for num_simulations in range(1, 10001):
    roll_1 = np.random.choice(dice_rolls)
    roll_2 = np.random.choice(dice_rolls)
    if (roll_1 == 6) & (roll_2 == 6) :
        count_hits += 1
    if (num_simulations % 10) == 0:
        y_data.append(float(count_hits) / float(num_simulations))
        x_data.append(num_simulations)

# Plot the figure
plt.plot(x_data,y_data)
```

According to the multiplication rule $P(A_6 \ and \ B_6) = P(A_6) \cdot P(B_6)= \frac{1}{6} \cdot  \frac{1}{6} = \frac{1}{36}$ = 0.0278 which is close to the number obtained from the simulation.

## Conditional probability

Let's introduce the concept of conditional probability with an example using the `family_college` dataset (Diez et al, 2019). The dataset contains a sample of 792 cases with two variables: `teen` and `parents`, and is summarized in the table below.

The `teen` variable indicates whether the teenager attended college or not. The `parents` variable indicates whether or not at least one of the teenager's parents completed a college degree. We want to explore if there is any relationship between teenagers attending college and whether or not one of their parents completed a degree.

The contingency table for the data is shown below:

| Teens / Parents       | Parent Degree   |Parent No Degree      |Total    |
| :-----------: |:-------: |:-------:|:-------:|
| Teen College       |231       |214      |445      |                               
| Teen Not           |49        |298      |347      |
| Total         |280       |512      |792      |



The image above depicts the dataset as follows:

- 38% of the sample represents those where neither parent nor child completed a college degree


- 29% of the sample represents those where both parent and child completed a college degree


- 27% of the sample represents those where the parents did not complete a college degree, but their child did


- 6% of the sample represents those where the parents did complete a college degree, but their child did not

Is there a connection between the education level of the parents and the teenager? Is a parent having a college degree associated with college attendance for their children?

*Analysis:*

- The probability that a random teenager from the dataset attended college and that at least one of the teens parents has a college degree is 231 out of 280 cases:


$$P (teen\ college\ given\ parents\ degree) = 231 / 280 = 0.825$$


- The probability that a random teenager from the dataset did not attend college and that at least one of the teens parents has a college degree is 49 out of 280 cases:
  

$$P (teen\ college\ given\ parents\ degree) = 49 / 280 = 0.175$$


- These are *conditional probabilities* because we computed the probability under a condition: a parent has a college degree


- Notice that the conditional probabilities add up to 1


*Conclusion*: The conditional probabilities show that there is a connection.

**Definitions:**

The conditional probability of the outcome of interest $A$ given condition $B$ is computed as the following:


$$ P(A\ |\ B)=\frac{P(A\cap B)}{P(B)} $$


**NOTE:** $P(A\cap B)$ is the intersection of $A$ and $B$, or in other words $A \ and \ B$. The $|$ means "given" &mdash; $P(A \ |\ B)$ is "the probability of $A$ given $B$ occurred."

Applying the general definition to our example:


$$ P(Teen\  College\ |\ Parent\ Degree)=\frac{P(Teen\  College\cap Parent\ Degree)}{P(Parent\ Degree)} $$

## Marginal and joint probabilities

- **Joint probability**: This is the probability of simultaneous outcomes for two or more variables or processes. For example, the probability that a child went to college when their parents did not.

$$
P (teen\ college\ and\ parents\ not) = P (teen\ college \cap parents\ not)= \frac{214}{792} = 0.27
$$  

- **Marginal probability**: If a probability is based on a single variable, it is a marginal probability. For example, probability based solely on the teen variable is a marginal probability.

$$
P (teen\ college) = P(teen\ college \cap Parent\ degree) + P(teen\ college \cap Parent\ not) = \frac{231}{792} + \frac{214}{792}=\frac{445}{792} = 0.56
$$

The name marginal comes from the fact that they are located in the margin of probability tables.

## The general multiplication rule

From the definition of conditional probability:


$$ P(A\cap B)=P(A\ |\ B) \cdot P(B) $$


When **events $A$ and $B$ are independent**, the probability of event $A$ is not impacted by the occurrence of event $B$ and vice versa, so the following applies:


$$ P(A\ |\ B)=P(A)\ \ \ \ and\ \ \  \ P(B\ |\ A)=P(B) $$


Thus, we can conclude that in the special case of independent events:


$$ P(A\cap B)=P(A) \cdot P(B) $$

It is important to understand the difference between independent and disjoint events. The difference is that disjoint events are mutually exclusive and never occur at the same time, so for disjoint events $P(A \ and \ B) = 0$. Independent events may happen simultaneously and they do not impact each other, $P(A \ and \ B) = P(A) \cdot P(B)$.

If we toss a coin we may get either heads or tails: these two events are disjoint and do not happen simultaneously. If we toss two coins we will have an outcome of two independent events, one of: $ \{H, H\}, \{H, T\}, \{T, H\}$, or $\{T, T\} $.

**NOTE**: The marginal probabilities defined above describe disjoint events.

## Sum of conditional probabilities

Let $A_1$, ..., $A_k$ represent all the disjoint outcomes for a variable or process $A$. Then if $B$ is an event, possibly for another variable or process, we have:


$$ P(A_1 \ | \ B) +  + P(A_k \ | \ B) = 1 $$


The rule for complements also holds when an event and its complement are conditioned on the same information:


$$ P(A \ | \ B) = 1 - P(A^c \ | \ B) $$

## Independence considerations in conditional probability

If two events are independent, then knowing the outcome of one should provide no information about the other. We can show this is mathematically true using conditional probabilities.

From the definition of conditional probability:


$$ P(A\cap B)=P(A\ |\ B) \cdot P(B) $$


When events $A$ and $B$ are independent, the probability of event $A$ is not impacted by event $B$ and vice versa:


$$ P(A\ |\ B)=P(A)\ \ \ \ and\ \ \  \ P(B\ |\ A)=P(B) $$


In the special case of independent events:


$$ P(A\cap B)=P(A) \cdot P(B) $$

**Example**

Consider the case of flipping 2 coins (we'll call them $X$ and $Y$). The output of one coin tells us nothing about the output of the other. Let's compute the probability that both coins are heads.


$$ P(X=Heads \ and \ Y=Heads)=P(X=Heads \cap Y=Heads)=P(X=Heads) \cdot P(Y=Heads) = \frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4} $$


If we use the formula for conditional probability, again we see that the probability that $Y$ is heads does not depend on the probability of $X$ (since $X$ and $Y$ are independent):


$$P(Y=Heads \ | \ X=Heads) = \frac{P(Y=Heads \cap X=Heads)}{P(X=Heads)}=\frac{P(Y=Heads) \cdot P(X=Heads)}{P(X=Heads)}=P(Y=Heads) = \frac{1}{2}$$

**Exercise 3**

Socks in a drawer (Diez et al, 2019, p. 114 Exercise 3.24). In your sock drawer, you have 4 blue, 5 grey, and 3 black socks. Half asleep one morning, you grab 2 socks at random and put them on. Find the probability you end up wearing:

1. 2 blue socks 
2. No grey socks
3. At least 1 black sock 
4. A green sock
5. Matching socks

```python
# YOUR SOLUTION HERE
'''
Blue = 4
Grey = 5
Black = 3
---------
Total = 12


==========
Q1
2 Blue Socks
P(Blue and Black) = P(Blue) x P(Black)
                    = 4/12 x 3/11
                    = 
'''
print('P(Blue and Black) = ', 4/12 * 3/11)

'''
Q2
0 Grey Socks
P(not Grey and not Grey) = P(not Grey) x P(not Grey)
                        = (12-5)/12 x (11-5)/11
                        = 7/12 x 6/11
'''
print('P(not Grey and not Grey) = ', 7/12 * 6/11)

'''
Q3
1 Black Sock

P(Black and Black) = P(Black) x P(Black)
                    = 3/12 x 2/11
                    
P(Black and not Black) = P(Black) x P(not Black)
                        = 3/12 x (11-2)/11
                        = 3/12 x 9/11
                        
P(not Black and Black) =  P(not Black) x P(Black)
                        = (12-3)/12 x 3/11
                        = 9/12 x 3/11                                          
'''
print('P(1 Black) = ', (3/12 * 2/11) + (3/12 * 9/11) + (9/12 * 3/11))

'''
Q4
1 Green Sock
P(G and G) = P(G) x P(G)
            = 0/12 x 0/11
P(not G and G) = 12/12 x 0/11
P(G and not G) = 0/12 x 11/11
P(1 G) = 0

'''
print('P(1 G) = ', 0)

'''
Q5
Matching Socks
P(Blue and Blue) = P(Blue) x P(Blue)
                = 4/12 x 3/11
P(Grey and Grey) = P(Grey) x P(Grey)
                = 5/12 x 4/11
P(Black and Black) = P(Black) x P(Black)
                    = 3/12 x 2/11
'''
print('P(Matching) = ', (4/12 * 3/11) + (5/12 * 4/11) + (3/12 * 2/11))
```

**Solution**

**Q1:**

Socks: blue 4, grey 5, black 3, and total = 4+5+3 = 12   

* Probability of first blue = 4/12
* Probability of second blue = 3/11 (outcome is without replacement)

Using the multiplication rule $P(A\ and\ B)$ for independent outcomes:

$$P(A\ and\ B) = P(A)\cdot P(B)$$


**Answer: Probability of two blues = 4/12 \* 3/11 = 0.0909**

**Q2:**


* Probability of first non grey = (12-5)/12 = 7/12
* Probability of second non grey = (11-5)/11 = 6/11

**Answer: Probability of two non greys = 7/12 \* 6/11 = 0.3182**

**Q3:**

* Probability of first black and second non-black = 3/12 \* 9/11
* Probability of first non-black and second black = 9/12 \* 3/11
* Probability of two blacks = 3/12 \* 2/11

Using addition rule:
$$P(One\ Black) = P(Black and Non-Black) + P(Non-Black and Black) + P(two black)$$

Probability of at least one black = 3/12 \* 9/11 + 9/12 \* 3/11 + 3/12 \* 2/11 = 5/11

Or, using the complement rule:

* Probability of two non blacks = 9/12 * 8/11

$$P(One\ Black) = 1- P(Non\ Blacks)$$



**Answer: Probability at least one black  = 1 - (9/12 \* 8/11) = 0.4545**

**Q4:**

**Answer: Probability of a green sock is 0, because there are no green socks in the drawer**

**Q5:**

* Probability of first grey = 5/12  
* Probability of second grey = 4/11
* Probability of two grey = 5/12 \* 4/11

Probability of matching socks is the disjoint outcome of independent events without replacement (two blues, two greys, two blacks). Applying the addition rule to coloured paired outcomes will give us the desired probability:

$$ P(two\ blues\ or\ two\ greys\ or\ two\ blacks) = P(two\ blues) + P(two \ greys) + P(two\ blacks)$$


Using probabilities from previous answers:

**Answer: Probability of matching socks = (4/12 \* 3/11) + (5/12 \* 4/11) + (3/12 \* 2/11) = 0.2879**
