# Linear Regression Analysis

Linear regression analysis is a powerful method for modeling the relationship between two or more variables. Initially, we will focus on simple linear regression, which deals with only two variables. Afterwards, we will present an extended form known as multiple linear regression.

## Simple linear regression

The objective of **regression analysis** is to estimate the value of a random variable given the value of an associated variable. When the association is assumed to be linear, we refer to this as linear regression. A linear regression follows the equation:


$$y = \beta_0 + \beta_1 x\,$$


where 
  
  - $x$ is the **predictor**, often called the independent or explanatory variable 
  - $y$ is the **response**, or dependent variable
  - $\beta_0$ is the **intercept**
  - $\beta_1$ is the **slope**, or regression coefficient

This set of parameter estimates defines what we refer to as a **regression model**. The slope $\beta_1$ describes the expected change of the $y$ variable if the explanatory variable $x$ is increased by one unit. The intercept $\beta_0$ describes the estimated value of $y$ when $x = 0$.

Sometimes, the linear regression is expressed as follows:


$$\hat{y} = \beta_0 + \beta_1 x$$


This is to highlight that $\hat{y}$ is an estimate of $y$ &mdash; not its observed value. The fitted value is just the estimate of $y$ and generally differs from the observed value. This fact is often expressed as follows:


$$\hat{y}_i = y_i + e_i\,$$ 


where $i$ denotes the i-th observation of ($x_i, y_i$) and $e_i$ is called the **residual**. Residuals express the difference between the predicted values $\hat{y}$ and the observed values $y$. In general, the better the regression model the closer to zero the residuals are.

## Visual interpretation

In a simple linear regression we are trying to find a line that fits a set of points in a plane defined by $x$ and $y$, so that the total distance between the points and the line is as small as possible. This line is called the **regression line**.

![Linear_regression.png](attachment:Linear_regression.png "Visual interpretation")

The intercept $\beta_0$ is equal to the distance between the origin and the intersection between the regression line and the $y$ axis. The slope $\beta_1$ is equal to the tangent of the angle $\alpha$ between the regression line and the $x$ axis: $\beta_1 = \tan\alpha$. When the slope $\beta_1$ equals $1$, $\alpha$ is equal to $45^\circ$.


## Underlying assumptions

The successful application of linear regression requires several assumptions regarding $x$ and $y$ to be satisfied. The most important of these are:

- **Linearity**: Which means that the independent variable $x$ and the dependent variable $y$ are linearly associated


- **Constant variance**: Of the response variable $y$ regardless of the value of the independent variable $x$


- **Independence of errors**: Which means that the errors $e$ are not correlated with each other.
 
For more details about the assumptions of linear regression analysis please see: https://en.wikipedia.org/wiki/Linear_regression#Assumptions

# Fitting a Linear Model

## Quick estimation of the parameters

Before we introduce the most common approach to linear regression, let's take a look at a way to perform a quick estimation of the regression parameters:

**Slope**:


$$\beta_1 = \frac{\sigma_y}{\sigma_x}\,R_{x,y}$$


**Intercept**:


$$\beta_0 = \bar{y} - \beta_1 \bar{x}$$

where
 - $\sigma_x$, $\sigma_y$ denote the standard deviations of $x$ and $y$ respectively
 - $\bar{x}$ and $\bar{y}$ are the means of $x$ and $y$ respectively
 - $R_{x,y}$ is the Pearson's coefficient of the correlation between $x$ and $y$
 
However, this approach to estimation of the regression parameters is rarely performed as more precise computational methods can be used instead. However, the above equations provide a good hint as to the association between the regression parameters and the basic statistical properties of $x$ and $y$.

## Ordinary least squares (OLS) method

When estimating values of $y$, we are looking to find values for $\beta_0$ and $\beta_1$ that minimize the residuals.

The most common practice is to find the line that minimizes the ***sum of the squared residuals*** &mdash; this technique is called **ordinary least squares (OLS).**

Mathematically this can be formulated as follows:


$$\hat{\beta} = argmin_{\beta}\, (\sum e^2_i)$$

where
  - $\widehat{\beta}$ denotes the estimate of the regression parameters (slope $\beta_1$ and intercept $\beta_0$)
  - $i$ is the index of the observation (i.e. pairs of $x_i$ and $y_i$ values)
  - $e_i$ is the residual associated with the $i$-th observation
  - $argmin_{\beta}$ denotes the procedure of searching for estimates of $\beta_0$ and $\beta_1$ that minimize the sum of the squared residuals

An important consequence of this approach is that it gives a higher weight to points with higher residuals (since the residuals are squared). Regression parameters (intercept and slopes) that are estimated using the OLS methodology are considered **"optimal."** The resulting regression line is often referred to as the **least-squares line.**

Using a small set of example data, we will:

1. Plot the data
2. Calculate the correlation between $x$ and $y$
3. Provide a quick estimate of the regression parameters $\beta_0$ and $\beta_1$
4. Use the OLS method to a generate a linear regression model
5. Review the regression model (draw the regression line, inspect the residuals)

```python
# Import the libraries first

import numpy as np
import pandas as pd

# We will use matplotlib to plot figures
import matplotlib.pyplot as plt

# For regression analysis we will use the statsmodels package
import statsmodels.formula.api as sm

# For visual inspection of the regression models
from statsmodels.graphics.regressionplots import plot_regress_exog, plot_fit, plot_leverage_resid2, influence_plot
```

```python
# Take set of arbitrary data
x = [1., 2., 3., 4. , 5., 6., 7., 9., 10.]
y = [1.3, 2.4, 3.1, 5.1, 3.8, 5.5, 9.1, 8.6, 9.3]
```

```python
# For convenience we will bind x, y into one data frame
d = pd.DataFrame(np.array([x, y]).T, columns= ['x', 'y'])

# Let's take a look at the data
d
```

```python
# And finally we will plot the data

plt.plot(d.x, d.y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Before we fit the actual regression model using the OLS method, we will try to quickly estimate the parameters $\beta_0$ and $\beta_1$ using the mean and standard deviation of $x$ and $y$ along with their correlation coefficient.

```python
# Let's calculate the Pearson's correlation coefficient first

# This will calculate the correlation matrix
cm = d.corr(method = 'pearson')

# Print correlation matrix
cm
```

The above matrix summarizes the correlations between $x$ and $y$.

**NOTE**: All values along the main diagonal of a correlation matrix are equal to one, as the correlation between any variable and itself is always 1.

```python
# Calculate the means of x, y

mu_x = np.mean(x)
mu_y = np.mean(y)

print(mu_x, mu_y)
```

```python
# Calculate the standard deviations of x, y

sig_x = np.std(x, ddof=1)
sig_y = np.std(y, ddof=1 )

print(sig_x, sig_y)
```

```python
beta_1 = sig_y/sig_x * cm.loc['x', 'y']
beta_0 = mu_y - beta_1 * mu_x

print(beta_0, beta_1)
```

We have estimated the intercept ($\beta_0$) to be approximately 0.50426, and the slope ($\beta_1$) to be approximately 0.92897.

Now we will fit a regression model using the OLS method.

```python
# Initialize the regression model

m = sm.ols('y ~ x', data = d)
```

```python
# Fit the regression model

m = m.fit()
```

```python
# Let's take a look at the regression parameters

m.params
```

As we can see, the values differ only slightly from our rough estimates. However, in general, this is rarely the case. Therefore, we shouldn't rely on the quick estimates and should always fit the model using OLS or a similar method.

Finally, we will visualize our results.

```python
# Calculate estimates of y

intercept, slope = m.params
d['y_est'] = intercept + slope * d.x
```

```python
# Let's plot the estimates along the actual observations

plt.plot(d.x, d.y, 'o', label = 'Observations')
plt.xlabel('x')
plt.ylabel('y')

# Draw estimates
plt.plot(d.x, d.y_est, 'r+', label='Estimates')

# Add legend
plt.legend()
plt.show()
```

```python
# As you may notice, all the estimates lie on the line;
# this is the regression line

plt.plot(d.x, d.y, 'o', label = 'Observations')
plt.xlabel('x')
plt.ylabel('y')

# Draw estimates
plt.plot(d.x, d.y_est, 'r+', label='Estimates')

# Plot regression line
plt.plot(d.x, d.y_est, 'r-', label='Regression line')

# Add legend
plt.legend()
plt.show()
```

```python
# We can even depict the errors associated with individual observations

plt.plot(d.x, d.y, 'o', label = 'Observations')
plt.xlabel('x')
plt.ylabel('y')

# Draw estimates
plt.plot(d.x, d.y_est, 'r+', label='Estimates')

# Plot regression line
plt.plot(d.x, d.y_est, 'r-', label='Regression line')

# Plot errors
plt.vlines(d.x, d.y_est, d.y, linestyles='dashed', label = 'errors')

# Add legend
plt.legend()
plt.show()
```

We can also take a look at the **model summary**, which provides a lot of useful information about our model. The meaning of these variables will be explained in the following sections.

```python
# Show the model summary

m.summary()
```

# How to Evaluate a Linear Model

Every regression model must be subjected to examination to evaluate its validity (i.e. its ability to provide meaningful estimates of $y$). Examination of a regression model typically involves multiple steps.

## Examination of regression residuals

The residuals of a valid regression model must be **(nearly) normally distributed and centered close to 0.** Failure to meet this requirement may indicate a misspecified model &mdash; i.e. the given explanatory variable is not sufficient to estimate $y$, or may indicate the presence of influential points (we'll discuss this later). If the mean is not zero the model is said to be **biased**.

If observations are independent &mdash; i.e. there is no underlying structure in the data &mdash; then the variability of the residuals must be constant across the whole dataset. 

A quick way to assess the residuals is to inspect them visually. The figures below show several examples of invalid regression caused by violation of the regression assumptions/criteria.

From left to right:

  - Relationship is not linear
  
  
  - Presence of influential points
  
  
  - Residual variability is not constant
  
  
  - Residuals are auto-correlated

![Visual_inspection.png](attachment:Visual_inspection.png "Evaluation by visual inspection")

Another visual technique to examine the model's residuals is the **normal probability plot**. The normal probability plot compares quantiles of the actual residuals against theoretical ones &mdash; i.e. quantiles of the normally distributed variable with zero mean, and standard deviation equal to the actual residuals.

Ideally, the residuals should be normally distributed and all the points on the plot should lie on the identity line (the line that forms a 45Â° angle with the x-axis). Deviations from the identity line suggest departures from normality.

![QQ_plot.png](attachment:QQ_plot.png "Normal probability plot example")

## Root mean square error

To assess the quality of the estimates of $y$, we often calculate the **root mean square error (RMSE)**:

$$\text{RMSE} = \sqrt{\frac{1}{N}\sum^N_i e^2_i}$$


where $N$ indicates the number of observations. 

To facilitate a comparison between datasets / models with different scales, it is useful to **normalize** the values of RMSE by dividing by the mean of $y$ ($\bar{y}$):

$$\text{NRMSE} = \frac{\text{RMSE}}{\bar{y}}$$


In some cases it is more desirable to normalize the values of RMSE by dividing by the range of the response variable ($y_{max} - y_{min}$):

$$\text{NRMSE} = \frac{\text{RMSE}}{y_{max} -\, y_{min}}$$

## Coefficient of determination

Another common way to assess a regression model is to calculate the correlation $R$ between the estimates $\widehat{y}$ and the actual observations $y$. Ideally the estimates and actual values of $y$ are perfectly correlated leading to a correlation coefficient equal to $1$ ($R = 1$). In practice this (almost) never happens. 

The correlation coefficient can take a range from $-1$ to $1$, where negative values indicate an inverse association between the estimates and the actual observations, which in practice happens only rarely (the model fit would have to be exceptionally bad). It is therefore more practical to use its square, called the **coefficient of determination**, or simply R-squared ($R^2$).

The $R^2$ provides a measure of how well the observed values are replicated by the regression model, based on the proportion of total variation of outcomes explained by the model. The values of R-squared range from 0 to 1, where $R^2 = 0$ indicates that there is no linear relationship between the model estimates and the observed values, while $R^2 = 1$ indicates that the model provides a perfect fit to the observed values.

$R^2$ can be calculated as:

$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$$


where $SS_{res}$ is the sum of squares of the model residuals $e$, often called the residual sum of squares; and $SS_{tot}$ is the total sum of squares: 

$$SS_{res} = \sum^N_i(\widehat{y}_i - y_i)^2$$


$$SS_{tot} = \sum^N_i(y_i - \bar{y})^2$$


In the case of linear regression, the total sum of squares can be easily partitioned into what is called explained variance ($SS_{explained}$) and the residual sum of squares ($SS_{res}$):

$$SS_{tot} = \sum^N_i(y_i - \bar{y_i})^2 = \sum^N_i(\widehat{y}_i + e_i - \bar{y})^2 = \sum^N_i(\widehat{y}_i - \bar{y} + e_i)^2 = \sum^N_i(\widehat{y}_i - \bar{y})^2 + \sum^N_i e^2_i = SS_{explained} + SS_{res}$$


From this we obtain:

$$R^2 = 1 - \frac{SS_{res}}{SS_{explained} + SS_{res}}$$


which implies that if:

- $R^2 = 1 \implies SS_{explained} = SS_{total} \implies SS_{res} = 0$ (i.e. the model explains all the variability in the observations)


- $R^2 = 0 \implies SS_{res} = SS_{total} \implies SS_{explained} = 0$ (i.e. the model explains nothing, no variability)

A formal proof of this can be found here: https://en.wikipedia.org/wiki/Partition_of_sums_of_squares

## Model summary

Most statistical software packages provide a standardized summary of regression analysis, typically containing several measures for evaluating the quality of the performed regression. Let's take a look at the output from the previous example:

| Parameter|value	|Parameter |value |
| --- |	--- | --- |	--- |
| **Dep. Variable:** | y | **R-squared:** | 0.895 |
| **Model:** | OLS | **Adj. R-squared:** | 0.880 |
| **Method:** | Least Squares | **F-statistic:** | 59.44 |
| **Date:** | Wed, 08 Aug 2018 | **Prob (F-statistic):** | 0.000115 |
| **Time:** | 12:00:31 | **Log-Likelihood:** | -12.056 |
| **No. Observations:** | 9 | **AIC:** | 28.11 |
| **Df Residuals:** | 7 | **BIC:** | 28.51 |
| **Df Model:** | 1 |		
| **Covariance Type:** | nonrobust |

The left side of the summary provides a description of the performed regression:

- **Dep. Variable** shows which data column served as the dependent variable (response)


- **Model** indicates the type of the regression analysis (in this case Ordinary Least Squares)


- **Method** is the method used to fit the regression parameters


- **Date/Time** is the date and time when the regression analysis was performed


- **No. Observations** is the number of observations


- **Df Residuals** is the number of residual degrees of freedom, defined as the number of observations minus one, minus the model's degrees of freedom (Df Model, below).


- **Df Model** is the number of degrees of freedom in the model, for simple linear regression this is always 1


- **Covariance Type** indicates whether robust regression analysis was applied (for more details visit: https://en.wikipedia.org/wiki/Robust_regression)
 
 
The right side of the summary provides an evaluation of the regression analysis (i.e. how well the model fits the data):

- **R-squared** is the coefficient of determination ($R^2$), which describes the amount of variation in the response that is explained by the model


- **Adj. R-squared** indicates the adjusted coefficient of determination. This is necessary when using multiple predictors, since any addition will improve the $R^2$ as the degrees of freedom decrease. This indicator balances the number of observations vs. the number of predictors. In practice, be cautious if $R^2$ and the adjusted $R^2$ differ greatly as this may indicate overfitting (to be discussed later).


- **F-statistic** is the value of the F-statistic (sometimes referred to as the F-value), that is calculated as:
 
$$F = \frac{SS_{explained}}{SS_{res}} \frac{n - k}{k - 1}$$
 
 
where $n$ is the total number of observations and $k$ is the number of degrees of freedom in the model.
 
- **Prob (F-statistic)** is often called the p-value, and indicates the probability that the true values of the slope and intercept are both equal to zero


- **Log-Likelihood** is the natural logarithm of the estimated maximum value of the **likelihood function** of the model. The likelihood function (or simply likelihood) is the probability that the true values of $beta$ are equal their estimated values $\hat{\beta}$ given the set of observations. We will discuss the likelihood function more thoroughly in the next module.


- **AIC** and **BIC** are the values of the Aikake information criterion (AIC) and the Bayesian Information Criterion (BIC). AIC and BIC are estimators of the relative quality of the statistical models for a given set of data, and they are calculated as follows:

$$\text{AIC} = 2k - 2\ln(\hat{L})$$


and 
 
$$\text{BIC} = k\ln(n) - 2\ln(\hat{L})$$


where $n$ is the total number of observations, $k$ is the number of degrees of freedom in the model and $\hat{L}$ is the estimated maximum of the likelihood function of the model. The lower the values of AIC and BIC, the greater the relative quality of the model.

## Testing regression parameters

For most statistical software packages, the regression parameters are automatically subjected to statistical testing where the null hypothesis is that the true values of these parameters are equal to a certain value which we refer to as the null value. Here null means a situation that is considered to be ordinary and typically is interpreted as a failure of the regression model. In most cases we are interested if the estimates are different from zero. In addition, most software packages provide an evaluation of the uncertainty of the obtained estimates, such as standard deviation and a 95% confidence interval.


| params| coef | std err | t | P > $|$t$|$ |	[0.025 | 0.975] |
| --- | --- | --- | --- | --- | --- | --- |
| **Intercept** | 0.5043 |	0.720 |	0.701 |	0.506 |	-1.197 |	2.206
| **x** | 0.9290 |	0.120 |	7.710 |	0.000 |	0.644 |	1.214


- **coef** is the estimated value of the regression parameters: $\hat{\beta_0}$ and $\hat{\beta_1}$


- **std err** is the standard error of the parameter estimates, often denoted as $s.e.(\hat{\beta_0})$ and $s.e.(\hat{\beta_1})$


- **t** is the value of the **t-statistic**, which is calculated as:
 
$$t_{\hat{\alpha}} = \frac{\hat{\alpha} - \alpha_0}{s.e.(\hat{\alpha})}$$
 
$$t_{\hat{\beta}} = \frac{\hat{\beta} - \beta_0}{s.e.(\hat{\beta})}$$
 
where $\alpha_0$ and $\beta_0$ are null values against which we are testing the estimates. In this case (and in most cases) they are equal to zero.

- **P > $|$t$|$** is the p-value associated with the given parameter &mdash; i.e. the probability that the true value of the given regression parameter is equal to the null value, given the set of observations. If the null value is equal to zero, the p-value of the slope can be interpreted as the probability that $x$ and $y$ are not linearly associated.


- The remaining two columns of the output show a **95% confidence interval** of the given regression parameter

# Outliers in Linear Regression

**Outliers** in a regression are observations that fall far from the "cloud" of ($x$, $y$) points.  These points are especially important because they can have a strong influence on the least squares line. Points that fall horizontally away from the center of the cloud tend to pull harder on the line, so we call them points with high **leverage**. If one of these high leverage points appears to invoke its influence on the slope of the regression line then we call it an **influential point**. Usually we can say that a point is influential if &mdash; had we fitted the line without it &mdash; the influential point would have been unusually far from the least squares line.

## Types of outliers

The figures below show examples of common types of outliers.

- **Figure 1**: Outlier far from the other points, though it only appears to slightly influence the line.


- **Figure 2**: Outlier on the right, though it is quite close to the least-squares line, which suggests that it isn't very influential.


- **Figure 3**: Point far away from the cloud, and this outlier appears to pull the least-squares line up on the right; examine how the line around the primary cloud doesn't appear to fit very well.


 - **Figure 4**: There is a primary cloud and then a small secondary cloud of four outliers. The secondary cloud appears to be influencing the line somewhat strongly, making the least-squares line fit poorly almost everywhere.
 
 
 - **Figure 5**: There is no obvious trend in the main cloud of points and the outlier on the right appears to largely control the slope of the line.
 
 
 - **Figure 6**: Outlier far from the cloud. However, it falls quite close to the least-squares line and does not appear to be very influential.

## A word of caution with outliers

**Don't ignore outliers when fitting a final model.** If there are outliers in the data, they should not be removed or ignored without a good reason. Whatever the final model is, it will not be very helpful if it ignores the most exceptional cases. Important information may be contained in the outliers.

The objective is not to get a good fit, it is to understand the process you are modeling and to create a useful model. By eliminating outliers you are missing out the opportunity to understand different scenarios you may encounter in the future.

# Multiple Linear Regression

In the previous section, we were trying to estimate the value of a random variable given the value of an associated variable, and that the relationship between them is assumed to be linear. But, what if the response variable is simultaneously connected to multiple predictors? In this situation, we may apply an extended form of linear regression, known as **multiple linear regression**. Multiple linear regression is a linear regression with more than one predictor.

In general we write the model as follows:

$$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2  + \ldots + \beta_n x_n $$


where, similar to simple linear regression:

  - $y$ is the **response**, or the dependent variable
  - $x_1, x_2,\ldots, x_n$ are the **predictors**, aften called the independent or explanatory variables 
  - $\beta_0$ is the **intercept** and $\beta_1, \beta_2, \ldots, \beta_n$ are the **regression coefficients**
  - $n$ is the number of predictors used &mdash; often referred to as a model's **degrees of freedom**

In addition to the assumptions discussed with respect to simple linear regression, application of multiple linear regression assumes that the **individual predictors are independent of each other**.

## How to evaluate multiple linear regression model

Similar to simple linear regression, we must inspect the model residuals first. Residuals must be approximately normally distributed with a mean close to zero. To assess the independence of the model residual, we may plot the dependence of the residuals on the values of each of the model predictors.

To evaluate the multiple linear regression model, we can apply the measures that were previously described, such as $R^2$ or RMSE. However, a regression model with more predictors will **always have a better $R^2$ and RMSE**. This is because the increased number of predictors increases the number of parameters that can be tuned, so that the obtained $\hat{y}$ estimates are closer to the true $y$ values.

To account for this fact, we may need to use the **adjusted $R^2$**,

$$R_\text{adj}^2 = 1 - (1 - R^2)\frac{n - 1}{n - k - 1}$$


where:

  - $R^2$ is the unadjusted coefficient of determination
  - $n$ is the number of observations
  - $k$ is the number of degrees of freedom in the model

Importantly, unlike adjusted $R^2$, there is no adjusted form of RMSE to account for the number of predictors.

## Collinearity and confounding variables

One of the most common complications associated with multiple linear regression is when two or more of the predictors are linearly correlated with each other. This is referred to as **collinearity** (or **multicollinearity**, if it involves more than two predictors).

- Collinearity is when two or more predictors are correlated


- The stronger the correlation between the predictors, the stronger the collinearity

**Collinearity is dangerous** as it causes the coefficient estimates to change erratically in response to even a small change in the data. There are several important consequences of this phenomenon:

- Estimates of the individual regression coefficients $\beta$ are inaccurate<br><br>

  - We can't assess how well the predictors predict the response variable<br><br>
  - We can't identify which predictors are redundant with respect to others<br><br>
  
- It does not reduce the predictive power of the model, within the sample dataset


- It causes **overfitting** (see following sections)

To avoid the complications caused by collinearity, we must be cautious with the selection of the predictors for our regression model. Ideally, we want our predictors to have:

- Strong linear correlation with the response
 
 
- Weak linear correlation between each other
 
We must always assess the correlations between the predictors and the response, as well as the correlations among the predictors.

Collinearity typically emerges due to a presence of **confounding variables**. A confounding variable is a variable that affects the response but also other independent variables (predictors). A confounding variable may be one of the observed predictors, but it may also be a **latent variable** (or hidden variable) that is not involved in the model.

## Stepwise linear regression

**Stepwise regression** is a heuristic for fitting multiple regression models, in which the choice of predictors is carried out an automated process. Stepwise regression is a sequential process, where in each step a variable is selected from the set of candidate explanatory variables and is considered as a predictor. A pre-specified criterion is used to determine whether to keep the given variable as a model predictor. Commonly, the criterion can be the F-statistic of the resulting regression model, the adjusted $R^2$, the Akaike Information Criterion, or the Bayesian Information Criterion, to name a few. The two most common stepwise regression approaches are:

  - **Forward selection** (sometimes called step-up)
  
  
  - **Backward selection** (sometimes called step-down)

### Forward selection

The forward selection method begins without any variables in the model. For each of the candidate variables, a simple linear regression model is estimated and the variable that produces the highest $R^2$ is selected (other criterion may be applied). In the next step, each of the remaining candidate variables are temporary added to the model. The variable that increases the resulting $R^2$ of the model the most is preserved permanently. At each step, we select the candidate variable that increases $R^2$ the most. The procedure stops adding variables when none of the remaining variables increase the model's $R^2$.

### Backward selection

The backward selection method starts with all candidate variables in the model. At each step, a variable is temporarily removed from the model. The variable whose removal increases the $R^2$ of the model the most is removed permanently. This process typically continues until no further $R^2$ improvement is achieved.

# Linear Regression with Categorical and Ordinal Variables

There are situations where the independent variable can only have a limited number of possible values. This kind of variable is called **categorical**. Typically, categorical variables describe certain qualitative properties. The set of values for categorical variables are called **levels**. 

Some examples of categorical variables and their levels:

  - Blood type: O, A, B, AB
  - Account type: Chequing, Saving, Money Market
  - Favourite food: Tacos, Burger, Sushi, etc.

A special (and the simplest) case of categorical variables are **binary variables**, which can only have two levels.

Some examples of binary variables and their levels:

  - Patient survival status: Alive, Deceased
  - Financial transaction: Legit, Fraud
  - Did you play tennis today?: Yes / No

Another special type of variables are **ordinal variables** &mdash; levels of which can be ordered. Ordinal variables can be considered as something "between" categorical and quantitative variables.

Some examples of ordinal variables and their levels:

  - Highest achieved education: Elementary &lt; High school &lt; College
  - Credit card type: Silver &lt; Golden &lt; Platinum
  - How do you like this movie?: "Don't like it" &lt; "It's OK" &lt; "Like it" &lt; "Like it a lot"

Importantly, in contrast to quantitative variables, we cannot quantify the difference between the individual levels of ordinal variables (e.g. we cannot safely say how much "better" college education is than high school education, nor how big is the difference between "I like the movie" and "It's OK"). 

Both categorical and ordinal variables can be used in linear regression, but only rarely can they be used directly. In most cases using the categorical predictors requires application of a special procedure called **encoding**. The most common types of encoding are **one hot encoding** for categorical variables and **polynomial encoding** for ordinal variables.

Encoding converts the categorical/ordinal variable into a set of nominal variables, which can be subsequently used in regression analysis.

## One hot encoding

**One hot encoding** is a procedure in which a categorical variable is converted into a set of $k$ binary variables (often called **dummy variables** or **dummies**) where $k$ is the number of levels of the categorical variable. Each resulting binary variable corresponds to one level of the categorical variable and indicates whether the value of the categorical variable equals the given level. For regression analysis purposes, the dummies are treated as nominal variables, with only two possible values: $0, 1$.


For example, let's encode a variable indicating the favourite food of a person. Suppose, each person's favourite food can be either Burgers, Tacos, or Sushi. This categorical variable has three levels, so we will use three dummy variables for its encoding:

| Favourite food | Tacos? | Burgers? | Sushi? |
| --- | --- | --- | --- |
| Tacos | 1 | 0 |	0 |	
| Burgers | 0 | 1 | 0 |
| Sushi | 0 | 0 | 1 |

## Polynomial encoding

**Polynomial encoding** is a procedure in which an ordinal variable is converted into a set of $k$ nominal variables (dummies) such that:

- The first dummy (**linear**) contains the first power of the order of the levels of the values of the original variable, the second dummy (**quadratic**) contains the second power of the order of the levels of the values of the original variable, and so forth.

Typically, the number of dummy variables required to encode the ordinal variable **is one less than the number of levels** of the original ordinal variable. However, it is often impractical to use more then 3 dummy variables (linear, square and cubic).

For example, let's encode the variable indicating the highest achieved education of the person. Suppose each person can have either accomplished Elementary school, High school, or College. We will treat this variable as ordered, since there is an apparent order to the three levels. To each level of the variable we can assign its place within the order:

 - Elementary - 1
 - High school - 2
 - College - 3

As the variable has three levels, we will use only two dummy variables. The first variable will store the first power of the level's order, the second will store the second power. This way we obtain the following encoding:

| Highest education | Linear | Square |
| --- | --- | --- |
| Elementary | 1 | 1 |	
| High school | 2 | 4 |
| College | 3 | 9 |
