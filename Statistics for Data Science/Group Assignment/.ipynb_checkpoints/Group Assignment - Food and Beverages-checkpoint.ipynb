```python
data_file = 'Tdata.xlsx'
food_jobs = pd.read_excel(data_file)
food_jobs['REF_DATE'] =  pd.to_datetime(food_jobs['REF_DATE'], format="%Y-%m")

df = food_jobs.copy()
df['REF_DATE'] =  pd.to_datetime(df['REF_DATE'], format="%Y-%m")
```

```python
food_jobs = food_jobs.loc[food_jobs["Activities"] == 'Food and beverage services']
food_jobs
```

```python
plt.figure(figsize=(20,15))
plt.scatter(food_jobs.REF_DATE, food_jobs.VALUE);

plt.title('Food and Beverages Jobs')
plt.xlabel('Quarters')
plt.ylabel('Number of jobs')
plt.show()
```

```python
# Remove post-covid data (outliers)
food_summer_jobs = food_jobs[food_jobs['REF_DATE'].dt.strftime('%m') == '07']
total_food_summer_jobs = food_summer_jobs.loc[food_summer_jobs["Activities"] == 'Food and beverage services']
total_food_summer_jobs_temp = total_food_summer_jobs.copy()
total_food_summer_jobs_preCovid = total_food_summer_jobs_temp.drop(total_food_summer_jobs_temp.loc[total_food_summer_jobs_temp['REF_DATE'] > '01-12-2019'].index)

plt.figure(figsize=(20,15))
plt.plot(total_food_summer_jobs_preCovid.REF_DATE, total_food_summer_jobs_preCovid.VALUE);

plt.title('Food and Beverages Jobs Pre-Covid (Summer)')
plt.xlabel('Quarters')
plt.ylabel('Number of jobs')
plt.show()

mu_food_total_summer = total_food_summer_jobs_preCovid.VALUE.mean();
sigma_food_total_summer = total_food_summer_jobs_preCovid.VALUE.std();
print('mu (food summer)', mu_food_total_summer)
print('std (food summer)', sigma_food_total_summer)
```

```python
# Spring Food Jobs
food_spring_jobs = food_jobs[food_jobs['REF_DATE'].dt.strftime('%m') == '04']
total_food_spring_jobs = food_spring_jobs.loc[food_spring_jobs["Activities"] == 'Food and beverage services']
total_food_spring_jobs_temp = total_food_spring_jobs.copy()
total_food_spring_jobs_preCovid = total_food_spring_jobs_temp.drop(total_food_spring_jobs_temp.loc[total_food_spring_jobs_temp['REF_DATE'] > '01-12-2019'].index)

plt.figure(figsize=(20,15))
plt.plot(total_food_fall_jobs_preCovid.REF_DATE, total_food_fall_jobs_preCovid.VALUE);
plt.title('Food and Beverages Jobs Pre-Covid (Spring)')
plt.xlabel('Quarters')
plt.ylabel('Number of jobs')
plt.show()

mu_food_total_spring = total_food_spring_jobs_preCovid.VALUE.mean();
sigma_food_total_spring = total_food_spring_jobs_preCovid.VALUE.std();
print('mu (food spring)' , mu_food_total_spring)
print('std (food spring)', sigma_food_total_spring)
```

```python
total_food_jobs = df.loc[df["Activities"] == 'Food and beverage services']

total_food_jobs_temp = total_food_jobs.copy()
total_food_jobs_preCovid = total_food_jobs_temp.drop(total_food_jobs_temp.loc[total_food_jobs_temp['REF_DATE'] > '01-12-2019'].index)
plt.figure(figsize=(20,15))
plt.plot(total_food_jobs_preCovid.REF_DATE, total_food_jobs_preCovid.VALUE);

mu_food_total_all = total_food_jobs_preCovid.VALUE.mean();
sigma_food_total_all = total_food_jobs_preCovid.VALUE.std();
print('mu food', mu_food_total_all)
print('std food', sigma_food_total_all)
zscore_food_all = (mu_food_total_summer - mu_food_total_all) / sigma_food_total_all
print('zscore', zscore_food_all)
p_food_all = stat.norm.cdf(zscore_food_all)
print ('p is ', p_food_all )
```

```python
n_food_all = len(total_food_jobs_preCovid.index)
print('n food All = ', n_food_all)

n_food_summer_sample = len(total_food_summer_jobs_preCovid.index)
print('n Summer Sample = ', n_food_summer_sample)

se_food = sigma_food_total_all/sqrt(n_food_summer_sample)
print('SE Summer Sample = ', se_food)

plt.figure(figsize=(12,10))
x = np.linspace(150, 200, 100)
plt.plot(x, stat.norm.pdf(x, loc = mu_food_total_all, scale = se_food))

plt.title("Food and Beverages Null hypothesis")
```

H0 mean <= mu_food_total_all

HA mean > mu_food_total_all

alpha = 0.1 

```python
summer_jobs = df[df['REF_DATE'].dt.strftime('%m') == '07']
total_summer_jobs = summer_jobs.loc[summer_jobs["Activities"]=='Food and beverage services']
total_summer_jobs_temp = total_summer_jobs.copy()
total_summer_jobs_preCovid = total_summer_jobs_temp.drop(total_summer_jobs_temp.loc[total_summer_jobs_temp['REF_DATE'] > '01-12-2019'].index)
```

```python
n_food_summer_sample = len(total_food_summer_jobs_preCovid.index)
print('n Food Summer Sample = ', n_food_summer_sample)

se_food_summer = sigma_food_total_all/sqrt(n_food_summer_sample)
print('SE Food Summer Sample = ', se_food_summer)

mu_food_total_summer = total_food_summer_jobs_preCovid.VALUE.mean();
sigma_food_total_summer = total_food_summer_jobs_preCovid.VALUE.std();

print('mu food summer', mu_food_total_summer)
print('std food summer', sigma_food_total_summer)

zscore_food_summer = (mu_food_total_summer - mu_food_total_all) / (sigma_food_total_all /sqrt(n_food_summer_sample))
print('Z score for food summer sample is ', zscore_food_summer)

p_food_summer = stat.norm.cdf(zscore_food_summer)
print ('p food is ', p_food_summer)

p_value_food_2 = stat.norm.cdf(mu_food_total_summer, loc = mu_food_total_all, scale = se_food_summer)
print('P-value 2 for given sample is ', p_value_food_2)
print('P-value for given sample is ', p_food_summer)

x_cr_food_summer = stat.norm.ppf(0.90, loc = mu_food_total_all, scale = se_food_summer)
print('Critical Value is ', x_cr_food_summer)
if p_food_summer < 0.9:
    print("Summer season does not produce more jobs than usual for the food industry")
else:
    print("Summer season is effective on creating jobs for the food industry")
    
```

```python
n_food_winter_sample = len(total_food_winter_jobs_preCovid.index)
print('n Food winter Sample = ', n_food_winter_sample)

se_food_winter = sigma_food_total_all/sqrt(n_food_winter_sample)
print('SE Food winter Sample = ', se_food_winter)

mu_food_total_winter = total_food_winter_jobs_preCovid.VALUE.mean();
sigma_food_total_winter = total_food_winter_jobs_preCovid.VALUE.std();

print('mu food winter', mu_food_total_winter)
print('std food winter', sigma_food_total_winter)

zscore_food_winter = (mu_food_total_winter - mu_food_total_all) / (sigma_food_total_all /sqrt(n_food_winter_sample))
print('Z score for food winter sample is ', zscore_food_winter)

p_food_winter = stat.norm.cdf(zscore_food_winter)
print ('p food is ', p_food_winter)

p_value_food_2 = stat.norm.cdf(mu_food_total_winter, loc = mu_food_total_all, scale = se_food_winter)
print('P-value 2 for given sample is ', p_value_food_2)
print('P-value for given sample is ', p_food_winter)

x_cr_food_winter = stat.norm.ppf(0.90, loc = mu_food_total_all, scale = se_food_winter)
print('Critical Value is ', x_cr_food_winter)
if p_food_winter < 0.9:
    print("winter season does not produce more jobs than usual for the food industry")
else:
    print("winter season is effective on creating jobs for the food industry")
    
```

```python
n_food_spring_sample = len(total_food_spring_jobs_preCovid.index)
print('n Food spring Sample = ', n_food_spring_sample)

se_food_spring = sigma_food_total_all/sqrt(n_food_spring_sample)
print('SE Food spring Sample = ', se_food_spring)

mu_food_total_spring = total_food_spring_jobs_preCovid.VALUE.mean();
sigma_food_total_spring = total_food_spring_jobs_preCovid.VALUE.std();

print('mu food spring', mu_food_total_spring)
print('std food spring', sigma_food_total_spring)

zscore_food_spring = (mu_food_total_spring - mu_food_total_all) / (sigma_food_total_all /sqrt(n_food_spring_sample))
print('Z score for food spring sample is ', zscore_food_spring)

p_food_spring = stat.norm.cdf(zscore_food_spring)
print ('p food is ', p_food_spring)

p_value_food_2 = stat.norm.cdf(mu_food_total_spring, loc = mu_food_total_all, scale = se_food_spring)
print('P-value 2 for given sample is ', p_value_food_2)
print('P-value for given sample is ', p_food_spring)

x_cr_food_spring = stat.norm.ppf(0.90, loc = mu_food_total_all, scale = se_food_spring)
print('Critical Value is ', x_cr_food_spring)
if p_food_spring < 0.9:
    print("spring season does not produce more jobs than usual for the food industry")
else:
    print("spring season is effective on creating jobs for the food industry")
    
```

```python
n_food_fall_sample = len(total_food_fall_jobs_preCovid.index)
print('n Food fall Sample = ', n_food_fall_sample)

se_food_fall = sigma_food_total_all/sqrt(n_food_fall_sample)
print('SE Food fall Sample = ', se_food_fall)

mu_food_total_fall = total_food_fall_jobs_preCovid.VALUE.mean();
sigma_food_total_fall = total_food_fall_jobs_preCovid.VALUE.std();

print('mu food fall', mu_food_total_fall)
print('std food fall', sigma_food_total_fall)

zscore_food_fall = (mu_food_total_fall - mu_food_total_all) / (sigma_food_total_all /sqrt(n_food_fall_sample))
print('Z score for food fall sample is ', zscore_food_fall)

p_food_fall = stat.norm.cdf(zscore_food_fall)
print ('p food is ', p_food_fall)

p_value_food_2 = stat.norm.cdf(mu_food_total_fall, loc = mu_food_total_all, scale = se_food_fall)
print('P-value 2 for given sample is ', p_value_food_2)
print('P-value for given sample is ', p_food_fall)

x_cr_food_fall = stat.norm.ppf(0.90, loc = mu_food_total_all, scale = se_food_fall)
print('Critical Value is ', x_cr_food_fall)
if p_food_fall < 0.9:
    print("fall season does not produce more jobs than usual for the food industry")
else:
    print("fall season is effective on creating jobs for the food industry")
x_cr_food = x_cr_food_fall
```

```python
plt.figure(figsize=(10,8))
x = np.linspace(150, 200, 100)

plt.plot(x, stat.norm.pdf(x, loc = mu_food_total_all, scale= se_food))
plt.plot(x, stat.norm.pdf(x, loc= mu_food_total_summer, scale= se_food))
plt.fill_between(x=np.arange(150, x_cr_food, 0.1), y1=stat.norm.pdf(np.arange(150, x_cr_food, 0.1), loc= mu_food_total_all, scale= se_food), facecolor='blue')

plt.vlines(x_cr_food, 0, 0.08 , colors='r', linestyles='dotted')
plt.vlines(mu_food_total_summer, 0, 0.08 , colors='y')
plt.vlines(mu_food_total_fall, 0, 0.08 , colors='y')
plt.vlines(mu_food_total_winter, 0, 0.08 , colors='y')
plt.vlines(mu_food_total_spring, 0, 0.08 , colors='y')

plt.vlines(x_cr_food, 0, 0.03 , colors='r', linestyles='dotted')
plt.vlines(x_cr_food, 0, 0.03 , colors='r', linestyles='dotted')

plt.text(x=mu_food_total_summer - 2, y= 0.08, s= "Summer")
plt.text(x=mu_food_total_winter - 1.5, y= 0.0775, s= "Winter")
plt.text(x=mu_food_total_fall - 1, y= 0.08, s= "Fall")
plt.text(x=mu_food_total_spring - 1.5, y= 0.0775, s= "Spring")
plt.text(x=x_cr_food - 2, y = 0.0822, s= 'Critical Value')
plt.title("Null hypothesis (Food and Beverages)")
```

From the analysis, none of the seasons are significant to generate jobs in the Food and Beverage Service industry. The critical value is <b>183</b>.
