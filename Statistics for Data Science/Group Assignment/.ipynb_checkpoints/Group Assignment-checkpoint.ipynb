```python
data_file = 'Tdata.xlsx'
food_jobs = pd.read_excel(data_file)
food_jobs['REF_DATE'] =  pd.to_datetime(food_jobs['REF_DATE'], format="%Y-%m")
```

```python
food_jobs = food_jobs.loc[food_jobs["Activities"]=='Food and beverage services']
food_jobs
```

```python
plt.figure(figsize=(20,15))
plt.scatter(food_jobs.REF_DATE, food_jobs.VALUE);

plt.title('Food and Beverages Jobs')
plt.xlabel('Quarters')
plt.ylabel('Number of jobs')
plt.show()
```

```python
# Remove post-covid data (outliers)
food_summer_jobs = food_jobs[food_jobs['REF_DATE'].dt.strftime('%m') == '07']
total_food_summer_jobs = food_summer_jobs.loc[food_summer_jobs["Activities"]=='Food and beverage services']

total_food_summer_jobs_temp = total_food_summer_jobs.copy()
total_food_summer_jobs_preCovid = total_food_summer_jobs_temp.drop(total_food_summer_jobs_temp.loc[total_food_summer_jobs_temp['REF_DATE'] > '01-12-2019'].index)

plt.figure(figsize=(20,15))
plt.scatter(total_food_summer_jobs_preCovid.REF_DATE, total_food_summer_jobs_preCovid.VALUE);


plt.xlabel('Quarters')
plt.ylabel('Number of jobs ')
plt.show()
```
